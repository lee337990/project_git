```python
import cv2
import os
import numpy as np

# 데이터를 저장할 리스트
data = []
size = (100, 100)

# 파일 목록 불러오기
filename = os.listdir('../project9/DATA/normal_cup_st/')

# 이미지 처리
for x in filename:
    path = '../project9/DATA/normal_cup_st/' + x
    img = cv2.imread(path)
    img = cv2.resize(img, size)

    b, g, r = cv2.split(img)
    clahe = cv2.createCLAHE(clipLimit=3, tileGridSize=(16, 16))
    enhanced = clahe.apply(g)

    # Canny 엣지 검출
    edges = cv2.Canny(enhanced, 150, 200)

    # 데이터를 1D 배열로 변환 후 리스트에 추가
    data.append(edges.flatten())

# 리스트를 Numpy 배열로 변환 (샘플 수 X 특징 수)
data_array = np.array(data)

# 공분산 행렬 계산 (이미지 간 공분산)
covariance_matrix = np.cov(data_array, rowvar=True)  # 샘플 간 공분산

# 상관계수 행렬 계산
correlation_matrix = np.corrcoef(data_array, rowvar=True)

# 결과 출력
print('공분산 행렬 크기 :', covariance_matrix.shape)
print('공분산 행렬:')
print(covariance_matrix)

# 상관계수 행렬에서 대각선 값(자기 자신과의 상관계수)은 제외하고,
# 상관관계가 높은 값만 찾기 위해 upper triangle만 추출
n = correlation_matrix.shape[0]
upper_triangle = np.triu(correlation_matrix, k=1)  # 대각선 위 값만 추출

# 가장 높은 상관관계를 가진 두 데이터의 인덱스를 찾기
max_corr_value = np.max(upper_triangle)  # 가장 높은 상관계수
max_corr_indices = np.unravel_index(np.argmax(upper_triangle), upper_triangle.shape)  # 인덱스 반환

# 가장 높은 상관관계를 가진 두 변수
print(f"가장 높은 상관계수: {max_corr_value}")
print(f"상관관계가 가장 높은 데이터 인덱스: {max_corr_indices}")

# 가장 높은 상관관계를 가진 두 이미지의 파일 이름 출력
image1_filename = filename[max_corr_indices[0]]
image2_filename = filename[max_corr_indices[1]]

print(f"가장 높은 상관관계를 가진 이미지 1: {image1_filename}")
print(f"가장 높은 상관관계를 가진 이미지 2: {image2_filename}")

# 가장 높은 상관관계를 가진 두 이미지를 불러오기
image1_path = '../project9/DATA/normal_cup/' + image1_filename
image2_path = '../project9/DATA/normal_cup/' + image2_filename

# 이미지 읽기
image1 = cv2.imread(image1_path)
image2 = cv2.imread(image2_path)

# 이미지 출력
cv2.imshow('Most Correlated Image 1', image1)
cv2.imshow('Most Correlated Image 2', image2)

# 이미지가 표시되고 ESC 키를 눌렀을 때 창 닫기
cv2.waitKey(0)
cv2.destroyAllWindows()
```
