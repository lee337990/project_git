```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

[연도별 지역별 공연 건수(2013~2024)]<hr>

```python
file1 = r'C:\Users\KDP-31-\Desktop\KDT7\math\MiniProject7\공연예술_공연건수_세분화_2013_2024_2.csv'
```

```python
df1 = pd.read_csv(file1)
```

```python
df1 = df1.drop(columns=['연도'])
```

```python
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']

df1.plot(kind='bar', stacked=False, figsize=(14,8))

plt.title('연도별 지역별 공연 건수 (2013~2024)', fontsize=16)
plt.xlabel('연도', fontsize=12)
plt.ylabel('공연 건수 [건]', fontsize=12)
plt.xticks(ticks=range(len(years)), labels=years, rotation=45)
plt.legend(title="지역", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 서울, 경기, 인천 데이터 합치기
df1['수도권'] = df1['서울'] + df1['경기'] + df1['인천']
```

```python
# 서울, 경기, 인천 제외한 지역 합치기
other_regions = ['부산', '대구', '경남', '경북', '울산', '대전', '대전', '충북', '충남', '세종', '광주', '전북', '전남', '강원도', '제주도']
df1['비수도권'] = df1[other_regions].sum(axis=1)
```

```python
# 꺾은선 그래프 그리기
plt.figure(figsize=(14,8))

df1['수도권'].plot(kind='line', marker='o', linestyle='-', label='수도권', color='b', figsize=(14, 8))
df1['비수도권'].plot(kind='line', marker='s', linestyle='--', label='비수도권', color='r', figsize=(14, 8))

plt.title('수도권(서울, 경기, 인천)과 비수도권 공연 건수 (2013~2024)', fontsize=16)
plt.xlabel('연도', fontsize=12)
plt.ylabel('공연 건수 [건]', fontsize=12)
plt.xticks(ticks=range(len(years)), labels=years, rotation=45)
plt.legend(title="구분", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 수도권(서울, 경기, 인천)의 공연 건수와 순이동 수 비교 (2013~2024)
import numpy as np

df1['수도권']
df4['수도권']

years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']

# 상관관계
correlation = df1['수도권'].corr(df4['수도권'])

print(f"수도권(서울, 경기, 인천) 공연 건수와 순이동 수의 상관계수: {correlation}")

# 꺾은선 그래프 그리기
fig, ax1 = plt.subplots(figsize=(14, 8))

# 1번 수도권 데이터 (왼쪽 y축)
ax1.plot(df1.index, df1['수도권'], color='b', marker='o', label='수도권 공연 건수', linestyle='-')
ax1.set_xlabel('연도', fontsize=12)
ax1.set_ylabel('공연 건수 [건]', fontsize=12, color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 공연 건수 추세선 추가
x = np.arange(len(years))
z1 = np.polyfit(x, df1['수도권'], 1)
p1 = np.poly1d(z1)
ax1.plot(df1.index, p1(x), color='skyblue', linestyle='dotted', label='공연 건수 추세선')

# 오른쪽 y축 설정
ax2 = ax1.twinx()
ax2.plot(df4.index, df4['수도권'], color='r', marker='s', label='수도권 순이동 수', linestyle='--')
ax2.set_ylabel('순이동 수 [명]', fontsize=12, color='r')
ax2.tick_params(axis='y', labelcolor='r')

# 순이동 수 추세선 추가
z2 = np.polyfit(x, df4['수도권'], 1)
p2 = np.poly1d(z2)
ax2.plot(df4.index, p2(x), color='orange', linestyle='dotted', label='순이동 수 추세선')

# 범례 추가
ax1.legend(loc='upper left')  # 왼쪽 y축 범례
ax2.legend(loc='upper right')  # 오른쪽 y축 범례

# 제목 및 레전드 설정
plt.title('수도권(서울, 경기, 인천)의 공연 건수와 순이동 수 비교 (2013~2024)', fontsize=16)
plt.grid()
fig.tight_layout()
plt.show()
```
