```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 기업 정보 데이터 (사원수, 설립연도)
company_info = {
    '엑소시스템즈': {'사원수': 18, '설립연도': 2017},
    '오픈서베이': {'사원수': 115, '설립연도': 2011},
    '데이터라이즈': {'사원수': 58, '설립연도': 2019},
    '브레이브모바일': {'사원수': 128, '설립연도': 2014},
    '윌로그': {'사원수': 51, '설립연도': 2021},
    '에너닷': {'사원수': 30, '설립연도': 2018},
}

# 평점 데이터 (기업별 평점)
ratings = {
    '엑소시스템즈': [4.5, 3.5, 4.3, 5, 4.3, 5],
    '오픈서베이': [3.6, 3.5, 3.2, 3.7, 3.5, 3.5],
    '데이터라이즈': [4.6, 4.4, 3.1, 4.3, 3.7, 4.4],
    '브레이브모바일': [3.5, 3.2, 3.7, 3.6, 3.2, 2.9],
    '윌로그': [3.5, 3.2, 3.9, 3.8, 3.4, 3.2],
    '에너닷': [3.8, 2.6, 4.2, 3.8, 3.1, 3.5],
}

# 데이터프레임 생성
df_info = pd.DataFrame(company_info).T  # 기업 정보 (사원수, 설립연도)
df_ratings = pd.DataFrame(ratings).T  # 기업 평점 데이터
df_ratings.columns = ['전체평점', '복지', '워라벨', '사내문화', '승진기회', '경영진']

# 두 데이터프레임 합치기
df = pd.concat([df_info, df_ratings], axis=1)

# 회사명 순서대로 x축에 맞추기
company_names = df.index

# 이중 y축 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 왼쪽 y축: 전체평점
ax1.set_xlabel('회사명', fontsize=12)
ax1.set_ylabel('전체평점', color='b', fontsize=12)
ax1.bar(company_names, df['전체평점'], color='b', alpha=0.6, label='전체평점')
ax1.tick_params(axis='y', labelcolor='b')

# 오른쪽 y축: 사원수
ax2 = ax1.twinx()  # 이중 y축
ax2.set_ylabel('사원수', color='r', fontsize=12)
ax2.plot(company_names, df['사원수'], color='r', marker='o', label='사원수', linestyle='-', linewidth=2)
ax2.tick_params(axis='y', labelcolor='r')

# 그래프 제목 추가
plt.title("사원수와 전체평점의 관계", fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
