```python
# 상승률 계산 (북구)
bukgu_percentage = ((bukguDF.iloc[0, 0:] - bukguDF.iloc[1, 0:]) / bukguDF.iloc[1, 0:]) * 100

# 상승률 계산 (중구)
jungu_percentage = ((junguDF.iloc[0, 0:] - junguDF.iloc[1, 0:]) / junguDF.iloc[1, 0:]) * 100

# 결과 출력
print("북구 상승률:\n", bukgu_percentage)
print('\n\n')
print("중구 상승률:\n", jungu_percentage)
```
